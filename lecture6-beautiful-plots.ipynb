## Beautiful plots

### Tufte's principles of excellence

See http://www.edwardtufte.com and his four books on the insightful presentation of statistical data:

* The visual display of quantitative information - http://www.edwardtufte.com/tufte/books_vdqi
* Envisioning information - http://www.edwardtufte.com/tufte/books_ei
* Visual explanations - http://www.edwardtufte.com/tufte/books_visex
* Beautiful evidence - http://www.edwardtufte.com/tufte/books_visex

All four are recommended reading. In a nutshell, Tufte's credo is that

> Graphical excellence is the well-designed presentation of interesting data. To achieve excellence, the data must be substantive; its statistical treatment must be accurate; and its graphical presentation must be designed for clarity, precision, and efficiency.

Tufte's _principles of design_ may be stated as follows:

1. Making intelligent and appropriate comparisons
2. Using data to explain causality, mechanisms, structure
3. Remembering that data is often intrinsically multivariate, and finding creative ways to "escape Flatland"
4. Integrating different modes of evidence by bringing words, tables, and figures together (no legends!). Galileo was [doing it already](http://www.edwardtufte.com/bboard/q-and-a-fetch-msg?msg_id=00004p)...
5. Documenting the provenance of data used for plots

Tufte makes specific suggestions on achieving these principles:

To make comparisons (1):

* Maximize data ink (the amount of ink used for data as opposed to graphical structure) and data density
* When possible use paper, which still has the best resolution
* Put related data side by side: use [small multiples](http://www.edwardtufte.com/bboard/q-and-a-fetch-msg?msg_id=0000hv) and [sparklines](http://www.edwardtufte.com/bboard/q-and-a-fetch-msg?msg_id=0001OR) (see http://pygal.org/sparks for Python)

To use data to explain (2):

* Avoid chartjunk (think Powerpoint chart "effects") and visual clutter (e.g., noisy gridlines, messy icons, clashing colors)
* Use bright colors sparingly for important things, prefer natural palettes

To integrate evidence (4):

* Treat illustrations as maps, including a scale reference, and annotating important points with their values
* In flowcharts, words can replace boxes, lines and arrows indicate links and causality
* Use sparklines as datawords

Above all, Tufte stresses graphical integrity:

* Represent numbers proportionally in graphics (i.e., avoid distorting relative sizes with perspective effects, represent quantities by length or area as appropriate)
* Label plots clearly and thoroughly
* Show data variation, not design variation (i.e., most "information" in a plot should relate to the data, not to decoration or inessential graphical structure)

See also:

* http://www.sealthreinhold.com/tuftes-rules
* Tufte's biting critique of PowerPoint: http://www.edwardtufte.com/bboard/q-and-a-fetch-msg?msg_id=0001yB

Let's try to do some of this with `matplotlib`. The default settings are stored in the `rcParams` dictionary, which can be modified easily with the `matplotlib.rc` function. 

```python
# for instance, when we prepare plots for publication,
# it's best to work at the final graph size
# (here 5in x 3in for a largish figure in a paper)
# then we can set the fontsizes at the size they take
# in the main text, perhaps 10 or 11 points
mp.rc('figure',figsize=[5,3])

mp.rc('axes',labelsize=12)
mp.rc('xtick',labelsize=10)
mp.rc('ytick',labelsize=10)
mp.rc('legend',fontsize=12)

# depending on the journal, we may want to use a font
# like Times New Roman, which now has a suitable mathematical
# counterpart in the STIX serif fonts
# mp.rc('font',family='Times New Roman',size=10)
# mp.rc('mathtext',fontset='stix')

# or maybe we want to use the TeX fonts
# mp.rc('font',family='Computer Modern Roman',size=10)
# mp.rc('mathtext',fontset='cm')

# or (perhaps for a presentation) Helvetica fonts
# for both text _and_ math
mp.rc('font',family='Helvetica',size=12)
mp.rc('mathtext',fontset='stixsans')

# 42 is TTF, required for Adobe Illustrator editing
mp.rc('pdf',fonttype=42)
```

Let's exercise `matplotlib` a little more, using the Matlab-like `matplotlib.pyplot` interface. We see how we restrict the axes, label them (using math expressions if desired), set custom tick locations and labels, annotate the plot with an arrow, and add shapes.

```python
# a pedestrian plot
fig = pp.figure()

x = np.linspace(0,4*math.pi,100)
y = np.sin(x)

# plot takes all the usual matlab options
pp.plot(x,y)

# restrict the axis to where we want it
pp.axis([0,4*math.pi,-1,1])

# labels --- in all text, feel free to mix in LaTeX expressions
pp.xlabel('$\phi$ (radians)')
pp.ylabel('amplitude')

# do our own ticks
ticklocations = [math.pi * i for i in range(1,5)]
ticklabels = [('$%s\pi$' % i) for i in range(1,5)]
pp.xticks(ticklocations,ticklabels)

# annotate points
dataxy = (x[40],y[40])
textxy = (0.5*math.pi,-0.5)
pp.annotate('note',dataxy,textxy,
            verticalalignment='center',horizontalalignment='center',
            arrowprops={'arrowstyle': '-|>','fc': 'k'})

# add geometric shapes... for that we need to grab
# the current "axes" first with gca()
art = mp.patches.Circle((math.pi,0),0.5,facecolor='gray',edgecolor='none')
pp.gca().add_patch(art);
```

The default `matplotlib` colors for plotlines are not ideal. We can replace them with one of the sets "brewed" according to Cynthia Brewer's guidelines, which distinguish between _sequential_ maps (good to display continuous ranges in maps), _diverging_ maps (e.g., for continuous ranges where we want to identify positive and negative), and _qualitative_ maps (to distinguish between lines most clearly).

* Her paper: http://www.personal.psu.edu/cab38/ColorSch/ASApaper.html
* A web visualization tool for color sets: http://colorbrewer2.org
* A Python package to use them in `matplotlib`: `brewer2mpl`

(With thanks to Olga Botvinnik for the example: http://nbviewer.ipython.org/gist/olgabot/5357268)

See also:

* https://jakevdp.github.io/blog/2014/10/16/how-bad-is-your-colormap/

In fact, `matplotlib` implements a set of [color maps](http://matplotlib.org/examples/color/colormaps_reference.html) that seem to follow Brewer's lead, but they're easier to use for continuous shading.

For instance, the standard "jet" map used in 2D histograms and density plots uses hue to encode value, so it creates spurious banded artifacts where none exist in the data.

```python
import matplotlib.colors as mc

pp.figure(figsize=(8,5))

x = np.random.randn(100000)
y = np.random.randn(100000) + 5

pp.hist2d(x,y,bins=40,norm=mc.LogNorm())
pp.colorbar()
```

It's much better to use a sequential map.

```python
themap = brewer2mpl.get_map('Blues','sequential',9).mpl_colormap

# or using the matplotlib version
# themap = pp.get_cmap('Blues')
```

```python
pp.figure(figsize=(8,5))

pp.hist2d(x,y,bins=40,norm=mc.LogNorm(),cmap=themap)
pp.colorbar()
```

It's easy to make new colormaps that interpolate between a few given colors, described as RGB.

```python
mymap = mp.colors.LinearSegmentedColormap.from_list('mymap',colors=[(0.85,0.85,0.85),(0.75,0.95,0.75),(0.85,0.85,0.85)])
```

...although that map does not make much sense in the context of this paper...

```python
pp.figure(figsize=(8,5))

pp.hist2d(x,y,bins=40,norm=mc.LogNorm(),cmap=mymap)
pp.colorbar()
```

Next, we'll do a Tuftian restyling of a standard bar plot.

```python
years = np.arange(2004, 2009)
print(years)
```

The x-axis ticks, which are oblivious to the notion of year, are especially silly.

```python
years = np.arange(2004, 2009)
heights = np.random.random(years.shape) * 7000 + 3000

pp.bar(years,heights)
```

In practice, we'd obtain the result below by iterative experimentation within iPython notebook, which is especially great to add options one by one and see the result.

```python
# change the colors of the bars, and remove the black lines
pp.bar(years,heights,
       linewidth=0,
       color=brewer2mpl.get_map('Set2', 'qualitative', 5).mpl_colors)

# add textual tick labels at just the right position (note the x offset)
pp.xticks(years+0.4,[str(year) for year in years])

# add the actual counts at the top of each bar
for year, value in zip(years,heights):
    pp.text(year+0.4,value + 160,'%d' % value,
            horizontalalignment='center')
    
# restrict the axis for best framing
pp.axis(xmin=2003.7, ymax=10500)

# remove the spines (using the object-oriented interface)
ax = pp.gca()
for d in ['top','bottom','right','left']:
    ax.spines[d].set_visible(False)
    
# finally, overlay a thin grid
pp.grid(axis='y',color='white',linestyle='-',linewidth=0.5)
```

```python
flip = 1
x = np.linspace(0, 14, 100)

for i in range(1, 7):
    pp.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
    
# but it's got its own functions, such as one to remove spines
seaborn.despine(left=True,bottom=True)
```

You can set the _style_ (among `darkgrid, whitegrid, dark, white, ticks}`), but also the context (among `paper, notebook, talk, poster`), which will change the size of the plot and the thickness of the lines.

```python
seaborn.set_context("talk")

for i in range(1, 7):
    pp.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

`seaborn` has also some value-added functions for statistical plots...

```python
seaborn.set_context("paper")

f, axes = pp.subplots(2, 2, figsize=(10,10), sharex=True)

rs = np.random.RandomState(10)
d = rs.normal(size=100)

b, g, r, p = seaborn.color_palette("muted",4)

seaborn.distplot(d, kde=False, color=b, ax=axes[0, 0])
seaborn.distplot(d, hist=False, rug=True, color=r, ax=axes[0, 1])
seaborn.distplot(d, hist=False, color=g, kde_kws={"shade": True}, ax=axes[1, 0])
seaborn.distplot(d, color=p, ax=axes[1, 1])

seaborn.despine(left=True)

pp.tight_layout()
```

```python
rs = np.random.RandomState(5)

mean = [0, 0]
cov = [(1, .5), (.5, 1)]

x1, x2 = rs.multivariate_normal(mean,cov,500).T

g = seaborn.jointplot(x1,x2,kind="scatter",size=6,space=0,ratio=10)
```

```python
g = seaborn.jointplot(x1,x2,kind="kde",size=6,space=0,ratio=10)
```
