## 1 Solving the Poisson equation with numpy

We want to solve Poisson's equation,

$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0$$

in the $[0,1] \times [0,1]$ square domain, with boundary conditions

$$u(x,0) = x, \quad u(x,1) = x - 1, \quad u(0,y) = -y, \quad u(1,y) = 1 - y.$$

We'll use the lowest-order finite-difference representation

$$\frac{\partial^2 u}{\partial x^2}(x_i,y_j)
\simeq \frac{1}{\Delta x} \left( \frac{\partial u}{\partial x}(x_{i+1},y_j) - \frac{\partial u}{\partial x}(x_{i-1},y_j) \right) \\
\simeq \frac{1}{\Delta x} \left( \frac{1}{\Delta x} \left( u(x_{i+1},y_j) - u(x_i,y_j) \right)
- \frac{1}{\Delta x} \left( u(x_{i},y_j) - u(x_{i-1},y_j) \right) \right)$$

leading to

$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \simeq \left( \frac{1}{\Delta^2}(u_{i+1,j} + u_{i-1,j} + u_{i,j+1} + u_{i,j-1} - 4u_{i,j}) \right)$$

### 1.1 Direct solution

The most direct approach is to solve the linear system directly:

```python
def boundary(grid):
    x = np.linspace(0,1,len(grid))
    
    grid[0,:]  = np.interp(x,[0,1],[0,1])
    grid[:,-1] = np.interp(x,[0,1],[1,0])
    grid[-1,:] = np.interp(x,[0,1],[-1,0])
    grid[:,0]  = np.interp(x,[0,1],[0,-1])
```

```python
def poisson_direct(gridsize,set_boundary):
    A = np.zeros(shape=(gridsize,gridsize,gridsize,gridsize),dtype='d')
    b = np.zeros(shape=(gridsize,gridsize),dtype='d')
    
    dx = 1.0 / (gridsize - 1)
    
    # discretized differential operator
    for i in range(1,gridsize-1):
        for j in range(1,gridsize-1):
            A[i,j,i-1,j] = A[i,j,i+1,j] = A[i,j,i,j-1] = A[i,j,i,j+1] = 1/dx**2
            A[i,j,i,j] = -4/dx**2
    
    # boundary conditions
    for i in range(0,gridsize):
        A[0,i,0,i] = A[-1,i,-1,i] = A[i,0,i,0] = A[i,-1,i,-1] = 1
    
    # set the boundary values on the right side
    set_boundary(b)
    
    return np.linalg.tensorsolve(A,b)
```

```python
sol = poisson_direct(25,boundary)
```

To show the solution, we need to bridge the usual interpretation of a matrix (row, column, from top and left) with the idea of plotting on the Cartesian plane.

### 1.2 Iterative solution

Let's try an iterative method: we will solve Poisson's equation by turning it into the diffusion equation

$$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}$$

and solving it to convergence by forward time-centered space differencing (FTCS)

$$u^{n+1}_{i,j} = u^{n}_{i,j} + \frac{\Delta t}{\Delta^2}(u^{n}_{i+1,j} + u^{n}_{i-1,j} + u^{n}_{i,j+1} + u^{n}_{i,j-1} - 4u^n_{i,j})$$

with the largest stable timestep $\Delta t = \Delta^2/4$ [1], leading to Jacobi's method

$$u^{n+1}_{i,j} = \frac{1}{4}(u^{n}_{i+1,j} + u^{n}_{i-1,j} + u^{n}_{i,j+1} + u^{n}_{i,j-1}),$$

which amounts to replacing each gridvalue with the average of its nearest neighbors (see Numerical Recipes ch. 19.5). This is consistent with the theory of harmonic functions.

[1] The _von Neumann stability analysis_ of the FTCS scheme is based on solving for the evolution of the _amplification factor_ $\xi(k)$ for modes of the form $u^n_j = \xi^n e^{i k j \Delta x}$ (see Numerical Recipes ch. 19.1).

```python
def jacobi(grid):
    newgrid = np.zeros(shape=grid.shape,dtype=grid.dtype)

    # apply evolution operator
    newgrid[1:-1,1:-1] = 0.25 * (grid[1:-1,:-2] + grid[1:-1,2:] +
                                 grid[:-2,1:-1] + grid[2:,1:-1])

    # copy boundary conditions
    newgrid[0,:]  = grid[0,:]
    newgrid[-1,:] = grid[-1,:]
    newgrid[:,0]  = grid[:,0]
    newgrid[:,-1] = grid[:,-1]
    
    return newgrid
```

We start from a random configuration, apply the boundary conditions, and iterate. We plot the solution intermittently. It converges.

```python
def initgrid(gridsize):
    x = np.random.randn(gridsize,gridsize)
    boundary(x)
    
    return x
```

```python
x = initgrid(25)

pp.figure(figsize=(12,12))

for i in range(300):    
    if i % 20 == 0:
        pp.subplot(4,4,i/20+1)
        showsol(x)
        pp.title('iter = %s' % i)
        
    x = jacobi(x)

pp.tight_layout()
```

It's actually more interesting to see the evolution of the error, which we define as the norm of the iterated solution minus the exact (direct) solution. We collect the errors in a list.

```python
x = initgrid(25)
sol = poisson_direct(25,boundary)

err_jacobi = []

for i in range(1000):    
    err_jacobi.append((i,np.linalg.norm(x - sol)))
            
    x = jacobi(x)

err_jacobi = np.array(err_jacobi)
    
pp.semilogy(err_jacobi[:,0],err_jacobi[:,1])
pp.title('Error after iteration: Jacobi averaging');
```

```python
%%timeit

x = initgrid(65)

for i in range(1000):    
    x = jacobi(x)
```

Let's see the alternative classical Gauss-Seidel method, which updates cells in sweeps, using new values as they become available. To implement this we need nested loops...

```python
def gauss_seidel(grid):
    newgrid = grid.copy()
    
    for i in range(1,newgrid.shape[0]-1):
        for j in range(1,newgrid.shape[1]-1):
            newgrid[i,j] = 0.25 * (newgrid[i,j+1] + newgrid[i,j-1] +
                                   newgrid[i+1,j] + newgrid[i-1,j])
    
    return newgrid
```

```python
x = initgrid(25)
sol = poisson_direct(25,boundary)

err_gs = []

for i in range(1000):    
    err_gs.append((i,np.linalg.norm(x - sol)))
            
    x = gauss_seidel(x)

err_gs = np.array(err_gs)
    
pp.semilogy(err_jacobi[:,0],err_jacobi[:,1])
pp.semilogy(err_gs[:,0],err_gs[:,1])

pp.title('Error after iteration: Jacobi and Gauss-Seidel averaging')
pp.legend(['Jacobi','Gauss-Seidel']);
```

Convergence speed is set by the largest eigenvalue of the iteration matrix, known as _spectral radius_. This is $1-\pi^2/(2J^2)$ for Jacobi, $1 - \pi^2/J^2$ for Gauss–Seidel.

For Jacobi, the number of iterations to reduce the error by a factor $10^{-p}$ is $\simeq p J^2/2$, while for Gauss–Seidel it's $p J^2/4$.

The Python loops slow down Gauss-Seidel considerably...

```python
%%timeit

x = initgrid(65)

for i in range(1000):    
    x = gauss_seidel(x)
```

### 1.3 Multigrid solution (worked out exercise)

In the _multigrid_ solution of this PDE, we're going to apply relaxation operators at different lengthscales, moving our solution down to coarser and up to finer resolutions using _restriction_ and _interpolation_ operators.

First we use matplotlib to plot nested grids... it will help us figure out the indexing.

```python
n = 5

x = np.arange(0,n)
x, y = np.meshgrid(x,x)

X = np.arange(0,n-0.5,0.5)
X, Y = np.meshgrid(X,X)

print "Coarser grid: %s; subgrid: %s" % (len(x), len(X))

ml.rcParams['xtick.major.pad'] = ml.rcParams['ytick.major.pad'] = 10

f, a = pp.subplots(figsize=(6,6))

a.plot(x,y,'ro',markersize=15)
a.plot(X,Y,'b.',markersize=10)

for s in a.spines:
    a.spines[s].set_color('none')

a.axis([-0.1,4.1,-0.1,4.1])

pp.savefig('multigrid.png')
```

Here's the setup of our nested grids.

```python
from IPython.display import Image

Image('./multigrid.png')
```

```python
# linear interpolation

def interpolate(grid):
    newlen = 2*len(grid)-1
    newgrid = np.zeros((newlen,newlen),dtype=grid.dtype)
    
    # copy common elements
    newgrid[0::2,0::2] = grid[:,:]
    
    # odd rows, odd columns
    newgrid[1::2,0::2] = 0.5*(grid[:-1,:] + grid[1:,:])
    newgrid[0::2,1::2] = 0.5*(grid[:,:-1] + grid[:,1:])
    
    # diagonal elements
    newgrid[1::2,1::2] = 0.25*(grid[:-1,:-1] + grid[:-1,1:] +
                               grid[1:,:-1]  + grid[1:,1:])
        
    return newgrid

# direct injection

def inject(grid):
    newlen = (len(grid) + 1)/2
    newgrid = np.zeros((newlen,newlen),dtype=grid.dtype)

    newgrid[:,:] = grid[::2,::2]
    
    return newgrid
```

We will now collect the sequence of operations that we wish to apply on the solution in a list; at the lowest resolution we want a full solution of the equation, so we create a function `repeat` that takes a function `func`, and returns a new function consisting of `func` iterated `n` times. This is an example of functional programming in Python.

```python
def repeat(func,n):
    def repeatedfunc(x):
        for i in range(n):
            x = func(x)
        return x
    
    return repeatedfunc
```

Remember that functions are first-class objects in Python, so there's nothing wrong with making a list of them. However, since variables are really labels, if we redefine a function we'll have to recreate the list, which would otherwise point to the _old_ function objects.

```python
opseq = [gauss_seidel_fast,
         inject,
         gauss_seidel_fast,
         inject,
         gauss_seidel_fast,
         inject,
         repeat(gauss_seidel_fast,20),
         interpolate,
         gauss_seidel_fast,
         interpolate,
         gauss_seidel_fast,
         interpolate,
         gauss_seidel_fast]
```

Try it out: it converges rapidly!

```python
x = initgrid(25)

pp.figure(figsize=(12,12))

for i,op in enumerate(opseq):    
    pp.subplot(4,4,i+1)
    showsol(x)
    pp.title('res = %s, op = %s' % (len(x),op.__name__))
        
    x = op(x)

pp.tight_layout()
```

This is ridiculously more effective than high-resolution iteration!

```python
x = initgrid(25)

sol = poisson_direct(25,boundary)

for op in opseq:
    x = op(x)

print np.linalg.norm(x-sol)
```

How much data at "CBCHIGH" categories 1 through 4? CBCHIGH denotes the search for high-mass binary inspirals. As for the categories,

* CAT1 = data that passes basic quality checks
* CAT2 = data that does not have obvious corruption or noisy features with clear correlations with instrument events and failure modes
* CAT3 = data that does not have even less obvious noisy features with unclear instrumental associations
* CAT4 = data that has been cleaned of glitchy features that result in excess triggers

A LIGO search would use CAT2 data; an upper-limit analysis would use CAT3.

```python
dqarray = quality['DQmask'].value

for cbchigh_cat in [1,2,3,4]:
    bitmask = 1 << cbchigh_cat
    print "%s seconds at CAT %d" % (sum(dqarray & bitmask > 0),cbchigh_cat)
```

We are going to work with CAT2 data. Let's plot which data is CAT2 in "raster" fashion, reshaping a vector to a 2D matrix. We'll use `matplotlib.pyplot.matshow` the latter. We want a discrete rather than continuous colormap (i.e., 1 = red, 2 = yellow, 3 = green, 4 = blue). Some quick googling shows that this can be achieved with `ListedColormap` and `BoundaryNorm`.

```python
dqplot = np.zeros(dqarray.shape)

for cbchigh_cat in [1,2,3,4]:
    bitmask = 1 << cbchigh_cat
    dqplot[dqarray & bitmask > 0] = cbchigh_cat
```

```python
import matplotlib.colors as mpc

cmap = mpc.ListedColormap(['red','yellow','green','blue'])
norm = mpc.BoundaryNorm([0,1,2,3],cmap.N)
    
pp.matshow(dqplot.reshape((16,256)),cmap=cmap,norm=norm,aspect=1.5);
```

We use numpy _masked arrays_ to obtain the segments of CAT2 data.
* http://docs.scipy.org/doc/numpy/reference/maskedarray.generic.html

```python
dq2 = (dqarray & (1 << 2)) > 0

masked = np.ma.masked_array(np.zeros(4096),~dq2)
segs = np.ma.flatnotmasked_contiguous(masked)
print segs
```

This will get us the strain data for segment `n`.

```python
def segdata(n):
    start, end = segs[n].start * 4096, segs[n].stop * 4096
    
    return ts[start:end], strain.value[start:end]
```

Let's look at the PSD for the first few segment, using `psd` from `matplotlib.mlab`.

```python
import matplotlib.mlab as mp
```

```python
def plotpsd(i):
    segt, segd = segdata(i)
    psd, freqs = mp.psd(segd,NFFT=2048,Fs=1.0/dt)

    pp.loglog(freqs[1:],np.sqrt(psd[1:]))
    pp.xlabel('$f/\mathrm{Hz}$'); pp.ylabel('$S(f)/\mathrm{Hz}^{1/2}$')
    pp.axis([10,1000,1e-24,1e-14]);
    pp.title('t = [%s,%s]' % (segt[0],segt[-1]))
```

To see if something goes "ping", we obtain a spectrogram using `matplotlib.mlab.specgram`. We examing segments of 4096 points = 1 sec.

We pick a prettier colormap than the default at http://matplotlib.org/examples/color/colormaps_reference.html

```python
def plotspec(i):
    segt, segd = segdata(i)
    pxx, freqs, t = mp.specgram(segd,NFFT=4096,Fs=1.0/dt)

    # normalize each frequency row by its median (effectively to whiten data) 
    pxx = np.array([row / np.median(row) for row in pxx])
    
    pp.pcolormesh(t,freqs,np.log10(pxx),cmap=pp.get_cmap('Oranges'))
    pp.xlabel('$t/\mathrm{s}$'); pp.ylabel('$S(f)/\mathrm{Hz}^{1/2}$')
    pp.axis([0,t[-1],freqs[0],freqs[-1]])
    pp.title('t = [%s,%s]' % (segt[0],segt[-1]))
```

```python
pp.figure(figsize=(12,16))

for s in range(16,24):
    pp.subplot(4,2,s-16+1)
    plotspec(s)

pp.tight_layout()
```
